# Duffing equation potential

Using the parameter values in the Wikipedia article.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pot(x):
    return -x**2/2. + x**4/4.
```

```python
x_pts = np.arange(-3.,3.,.01)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x_pts, pot(x_pts))
ax.set_xlim(-2.,2.)
ax.set_ylim(-.5,2.)
ax.set_xlabel('x')
ax.set_ylabel(r'$V(x) [{\rm Duffing}]$')
```
