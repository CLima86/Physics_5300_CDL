# Generating a triangular waveform with Scipy signal

We can use scipy.signal.sawtooth (https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.sawtooth.html#scipy.signal.sawtooth) to create an array with specified properties.

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
tau = 2.          # period of waveform
offset = tau/2.   # so there is a peak at t = 0
t_pts = np.arange(-10, 10.001, .001)
# width=0.5 generates triangular
triangle_pts = signal.sawtooth(2*np.pi * (t_pts - offset) /tau , width=0.5)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(t_pts, triangle_pts)
ax.set_xlim(-5,5)
ax.set_xlabel('t')

fig.tight_layout()
```
